```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(data={'Name': ['Joao', 'Allan', 'Trinity', 'Neo'], 'School':['Marista', 'Harvard', 'NYU', 'Oxford'], 'Grade': [70, 95, 98, 93]}, index=['Level1', 2, 'Level3', 4])
```

```python
df
```

```python
df['Grade_rev'] = df['Grade']/10
df
```

```python
df2 = pd.DataFrame(data={'Name': ['Joao', 'Allan', 'Trinity', 'Neo'], 'City': ['Maringa', 'NY', 'San Francisco', 'London']})
df2
```

```python
df = pd.merge(df, df2, on='Name', how='left')
df
```

```python
df.shape
df = df[['Name', 'City', 'School', 'Grade', 'Grade_rev']]
df.sort_values('Grade', ascending=True, inplace=True)
df
```

```python
df['Class'] = ['Overachiever' if x<=8 else 'Achiever' for x in df['Grade_rev']]
df
```
